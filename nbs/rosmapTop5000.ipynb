## For ISF

WGCNA for the top 5000 varied genes for each tissue.

- import the script

```r
source("/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/old_scripts/ClusteringBuilding.R")
```

- select expression files per tissue
- select tissues names

```r
rosmap_files <- c(
    "/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/ROSMAP_fixed_AC.csv",
    "/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/ROSMAP_fixed_MFBA9BA46.csv",
    "/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/ROSMAP_fixed_PCGBA23.csv"
)
```

```r
tissues = c("AC","MFBA9BA46","PCGBA23")
```

- Run the script with parameter for autobeta and top 5000 varied genes.

```r
XWGCNA_Clusters_autoBeta(
    tissue_names = tissues,
    tissue_expr_file_names = rosmap_files,
    sd_quantile = 0.0,
    max_genes_per_tissue = 5000,     
    out_prefix = "xwgcna_rosmap_autobeta_top5000",
    auto_beta = TRUE,
    plot_beta_curves = TRUE,
    blockwise_TOM = TRUE,
    TS_power = 3,
    CT_power = 2,
    group = "xwgcna_rosmap_autobeta_top5000",
    beta_method = "wgcna_new",
    save_intermediates = TRUE,
    wgcna_powerVector = seq(0.5, 20, length.out = 20),
    ct_fisher = FALSE,
    ct_fisher_scheme = "to_ref",
    ct_fisher_Nref = "max",
    ct_too_few_action = "zero",
    adj_type = "new"
)
```

- save csv files of adjecency matrix and tom.

```r
# load rds files
adj_matrix <- readRDS("/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/nbs/xwgcna_rosmap_autobeta_top5000_adjacency.rds")
tom_matrix <- readRDS("/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/nbs/xwgcna_rosmap_autobeta_top5000_TOM.rds")

# save as csv
write.csv(adj_matrix, "/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/nbs/xwgcna_rosmap_autobeta_top5000_adjacency.csv")
write.csv(tom_matrix, "/media/psylab-6028/DATA/Eden/CoExpression_ReProduction/nbs/xwgcna_rosmap_autobeta_top5000_TOM.csv")
```
